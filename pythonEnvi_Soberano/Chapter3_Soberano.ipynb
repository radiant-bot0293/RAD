Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		1.75
    [ 82,  85)		2.0
    [ 78,  81)		2.25
    [ 75,  77)		2.5
    [ 0,   74)		5.0

Adapted from [PEW] E3.3
<br>

```python
try:
    score_input = input("Enter a score between 0 and 100: ")
    score = float(score_input)


    if score < 0 or score > 100:
        print("Error: Score must be between 0 and 100.")
    else:
        if score >= 98:
            grade = 1.0
        elif score >= 94:
            grade = 1.25
        elif score >= 90:
            grade = 1.5
        elif score >= 86:
            grade = 1.75
        elif score >= 82:
            grade = 2.0
        elif score >= 78:
            grade = 2.25
        elif score >= 75:
            grade = 2.5
        else:
            grade = 5.0

        print(f"Grade: {grade}")

except ValueError:
    print("Error: Please enter a number.")
```
